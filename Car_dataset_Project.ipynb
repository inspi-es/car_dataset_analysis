```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Dataset**

```python
#loading a csv file
df_car=pd.read_csv("data.csv")

```

```python
df_car.head()
```

# **1. Data Cleaning**


###**1. Handle Missing Data:**


```python
#Identify if there are missing values
print(df_car.isnull().sum())
```

```python
sns.heatmap(df_car.isnull(), cbar=False, cmap="YlOrRd")
plt.title("Missing Data Heatmap")
plt.show()
```

```python
df_car['Market Category'] = df_car['Market Category'].fillna(df_car['Market Category'].mode()[0])
```

```python
df_car['Engine Fuel Type'] = df_car['Engine Fuel Type'].fillna(df_car['Engine Fuel Type'].mode()[0])
```

```python
df_car['Engine HP'] = df_car['Engine HP'].fillna(df_car['Engine HP'].median())
```

```python
df_car['Engine Cylinders'] = df_car['Engine Cylinders'].fillna(df_car['Engine Cylinders'].median())

```

```python
print(df_car.isnull().sum())
```

### **2. Data Type Conversion:**

```python
print(df_car.dtypes)
df_car.head()
```

```python
# Ensure 'Year' is integer
df_car['Year'] = pd.to_numeric(df_car['Year'], errors='coerce').astype('Int64')
```

```python
# Convert 'Number of Doors' to integer
df_car['Number of Doors'] = pd.to_numeric(df_car['Number of Doors'], errors='coerce').astype('Int64')
```

```python
df_car['Engine HP'] = pd.to_numeric(df_car['Engine HP'], errors='coerce')
df_car['Engine Cylinders'] = pd.to_numeric(df_car['Engine Cylinders'], errors='coerce')
```

```python
print(df_car.dtypes)
df_car.head()
```

### **3. Filter Data:**

```python
#The dataset to only include cars from the year 1995 and later.
df_car_filtered = df_car[df_car['Year'] >= 1995]
df_car_filtered.to_csv("cars_1995_and_later.csv", index=False)
df_car_filtered.head()
```

### **4. String Operations:**

```python
#Standardize text entries by converting these columns' entries to lowercase:
#Vehicle Style
df_car['Vehicle Style'] = df_car['Vehicle Style'].str.lower()

#Market Category
df_car['Market Category'] = df_car['Market Category'].str.lower()

df_car.to_csv("cars_lowercase.csv", index=False)
df_car.head()
```

# **2. Feature Engineering**


### **1. Create New Columns:**

```python
#Total MPG that is the average of city mpg and highway MPG
df_car['Total MPG'] = (df_car['city mpg'] + df_car['highway MPG']) / 2
df_car.to_csv("cars_with_total_mpg.csv", index=False)
df_car.head()
```

```python
#Create a column called Price per HP calculated as:
#MSRP / Engine HP.
df_car['Price per HP'] = df_car['MSRP'] / df_car['Engine HP']
df_car.to_csv("cars_with_price_per_hp.csv", index=False)
df_car.head()
```

# **3. Exploratory Data Analysis (EDA)**



### **1. Descriptive Statistics:**

```python
#Summary statistics (mean, median, standard deviation) for the following columns:
columns = ['Engine HP', 'MSRP', 'Popularity', 'highway MPG', 'city mpg']
eda_summary = pd.DataFrame({
    'Mean': df_car[columns].mean(),
    'Median': df_car[columns].median(),
    'Standard Deviation': df_car[columns].std()
})
print(eda_summary)
```

### **2. Group Analysis:**

```python
#Group the data by the following columns and calculate the average MSRP and Popularity for each group:
#Driven_Wheels
#Vehicle Size
#Engine Cylinders
grouped_summary = (
    df_car.groupby(['Driven_Wheels', 'Vehicle Size', 'Engine Cylinders'])
          .agg({'MSRP': 'mean', 'Popularity': 'mean'})
          .reset_index()
)
grouped_summary.rename(columns={'MSRP': 'Average MSRP', 'Popularity': 'Average Popularity'}, inplace=True)
grouped_summary
```

### **3. Visualizations**

```python
#Histogram that shows a distribution for the city mpg column.
plt.figure(figsize=(8, 5))
sns.histplot(df_car['city mpg'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of City MPG')
plt.xlabel('City MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
#Bar chart showing the average MSRP for each category in Vehicle Size.
avg_msrp_size = df_car.groupby('Vehicle Size')['MSRP'].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(x='Vehicle Size', y='MSRP', data=avg_msrp_size, palette='viridis')
plt.title('Average MSRP by Vehicle Size')
plt.ylabel('Average MSRP')
plt.xlabel('Vehicle Size')
plt.xticks(rotation=30)
plt.show()
```

```python
#Scatter plot showing the relationship between Engine HP and MSRP.
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Engine HP', y='MSRP', data=df_car, alpha=0.6)
plt.title('Engine HP vs MSRP')
plt.xlabel('Engine HP')
plt.ylabel('MSRP')
plt.show()
```

```python
#Boxplot showing the distribution of MSRP for each category in Driven_Wheels.
plt.figure(figsize=(8, 5))
sns.boxplot(x='Driven_Wheels', y='MSRP', data=df_car)
plt.title('MSRP Distribution by Driven Wheels')
plt.xlabel('Driven Wheels')
plt.ylabel('MSRP')
plt.xticks(rotation=30)
plt.show()
```

```python
#Line plot showing trends in average city mpg and highway MPG for different Transmission Type.
avg_mpg_transmission = df_car.groupby('Transmission Type')[['city mpg', 'highway MPG']].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.lineplot(x='Transmission Type', y='city mpg', data=avg_mpg_transmission, label='City MPG')
sns.lineplot(x='Transmission Type', y='highway MPG', data=avg_mpg_transmission, label='Highway MPG')
plt.title('Average City MPG & Highway MPG by Transmission Type')
plt.ylabel('MPG')
plt.xticks(rotation=30)
plt.legend()
plt.show()
```

# 4. Correlation Analysis:

```python
#Investigate the correlation between the following variables:
#Engine HP, MSRP, Popularity, city mpg, and highway MPG.
corr_columns = ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']
corr_matrix = df_car[corr_columns].corr()
print("Correlation Matrix:")
print(corr_matrix)

# Visualize correlation matrix as heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
